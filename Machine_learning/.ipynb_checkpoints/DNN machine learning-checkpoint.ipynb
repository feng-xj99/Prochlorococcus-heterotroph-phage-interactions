# Data standardization

```python
# load library
import os
import pandas as pd
import numpy as np
```

```python
import os
input_file = os.getenv('INPUT_FILE', 'default.txt')  # 默认为 default.txt 如果没有设置环境变量
cont = pd.read_table(input_file)
```

```python
data_full=cont.iloc[:,2:].astype(float)
target_full=np.array(cont.iloc[:,1]).astype(float)
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean')
data_full_feed = imp.fit_transform(data_full)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
data_full_std = scaler.fit_transform(data_full_feed)
```

```python
#Split the data
from sklearn.model_selection import train_test_split
data_train,data_test,target_train,target_test = train_test_split(data_full,target_full,test_size=0.3) #test size = 10%
```

```python
#Screen out the feature of p<0.05 in data_train
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd

# Initialize an empty resulting DataFrame
result = pd.DataFrame(columns=["Variable", "p_value"])

# run a t-test for each variable
for col in data_train.columns:
    # extract data from two groups
    group1 = data_train[target_train == 0][col]
    group2 = data_train[target_train == 1][col]

    # Perform a t-test and record the p-value
    t_stat, p_val = ttest_ind(group1, group2, nan_policy='omit')

    # Add results to result DataFrame
    result = result.append({"Variable": col, "p_value": p_val}, ignore_index=True)

result_name=(result['Variable'][result['p_value']<0.05]).values

data_train=data_train.loc[:,result_name]
data_test=data_test.loc[:,result_name]
```

```python
result = pd.concat([result, pd.DataFrame([{"Variable": col, "p_value": p_val}])], ignore_index=True)
```

# Tensorflow Keras 

```python
from tensorflow import keras
from keras import layers
from keras import regularizers
```

```python
#model definition
model = keras.models.Sequential()
```

```python
#input layer
model.add(layers.Flatten(input_shape=(data_train.shape[1:])))
```

```python
#number of layers, number of neurons
model.add(layers.Dense(8,activation="relu"))
model.add(layers.Dense(8,activation="relu"))
model.add(layers.Dense(6,activation="relu"))
model.add(layers.Dropout(0.5))
#model.add(layers.Dense(5,activation="relu"))
#model.add(layers.Dropout(0.5))
#model.add(layers.Dropout(0.5))
# model.add(layers.Dense(4,activation="relu"))
# model.add(layers.Dropout(0.5))
#model.add(layers.Dropout(0.5))
```

```python
#output layer
model.add(layers.Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
history=model.fit(data_train, target_train, epochs=100,validation_split=0.2,batch_size=32) #epochs가 배운 순환횟수
```

# Model performance evaluation

```python
def interval(model,test_x,test_y,target='Accuracy',interval_num=0.95):
    '''Put the desired indicator inside target= Example: AUC, Accuracy, Sensitivity, Specificity, Precision, F1'''
    from scipy import stats
    from sklearn.metrics import roc_curve,roc_auc_score
    from sklearn.metrics import auc
    from sklearn.metrics import confusion_matrix
    y_pred=(model.predict(test_x) > 0.5).astype("int32")[:,0]
    y_pred_proba=model.predict(test_x)[:,0]
    n_bootstraps = 1000
    rng_seed = 42  # control reproducibility
    bootstrapped_scores = []
    rng = np.random.RandomState(rng_seed)
    for i in range(n_bootstraps):
        # bootstrap by sampling with replacement on the prediction indices
        indices = rng.randint(0, len(y_pred), len(y_pred))
        if len(np.unique(test_y[indices])) < 2:
            # We need at least one positive and one negative sample for ROC AUC
            # to be defined: reject the sample
            continue
        TN, FP, FN, TP = confusion_matrix(test_y[indices], y_pred[indices]).ravel()
        Accuracy = (TP+TN)/(TP+FP+FN+TN)
        Sensitivity = TP/(TP+FN)
        Specificity = TN/(TN+FP)
        Precision = TP/(TP+FP)
        F1 = 2*(Sensitivity * Precision) / (Sensitivity + Precision)
        AUC = roc_auc_score(test_y[indices], y_pred_proba[indices])
        bootstrapped_scores.append(locals()[target])
    socres=np.array(bootstrapped_scores)
    mean=socres.mean()
    std=socres.std()/np.sqrt(len(socres))
    interval=stats.norm.interval(interval_num,mean,std)
    interval_num*100
    print('%svalue:'%target,mean)
    print('%s Starndard Error:'%target,std)
    print('%s'%(int(interval_num*100)),'%confidence interval:',interval)
```

```python
a1=['AUC','Accuracy','Sensitivity','Specificity','Precision','F1']
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 假设 model、data_test 和 target_test 已经定义
y_pred_keras = model.predict(data_test).ravel()

# 计算 ROC 曲线和 AUC
fpr, tpr, threshold = roc_curve(target_test, y_pred_keras)
roc_auc = auc(fpr, tpr)

# 绘制 ROC 曲线
plt.figure(figsize=(8, 6))  # 设置图形大小
plt.plot(fpr, tpr, 'b', label='Keras AUC = %0.2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')  # 绘制对角线
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.grid()  # 添加网格线

# 保存图形为文件
plt.savefig('ROC.curve.svg', format='svg', dpi=300, bbox_inches='tight')

# 显示图形
plt.show()
```
